```
def myLinSpace(firstPoint,lastPoint,numberOfTerms) :

  interval=(lastPoint-firstPoint)/(numberOfTerms-1) # formula used for finding the required interval between two numbers
  l=[] # list to store the numbers

  for i in range(numberOfTerms) :
    l.append(firstPoint+interval*i) # storing values
  
  return l

firstPoint=float(input("Enter First Point : ")) # accepting first point
lastPoint=float(input("Enter Last Point : ")) # accepting last point
numberofTerms=int(input("Enter How Many Terms : ")) # accepting number of terms to generate

l=myLinSpace(firstPoint,lastPoint,numberofTerms) # calling method
print(l) #showing resultant list
```
